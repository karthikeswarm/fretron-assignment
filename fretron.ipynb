```python
import matplotlib.pyplot as plt

# Corrected input coordinates (as per the visual pattern in the image)
flights = [
    ((1, 1), (2, 2), (3, 3)),
    ((1, 1), (2, 4), (3, 2)),
    ((1, 1), (4, 2), (3, 4))
]

# Plotting the paths
fig, ax = plt.subplots()
colors = ['red', 'blue', 'green']
labels = ['Flight 1', 'Flight 2', 'Flight 3']

for idx, flight in enumerate(flights):
    x_values = [point[0] for point in flight]
    y_values = [point[1] for point in flight]
    ax.plot(x_values, y_values, label=labels[idx], color=colors[idx])
    ax.scatter(x_values, y_values, color=colors[idx])

ax.legend()
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_title('Flight Paths')

plt.grid(True)
plt.show()
```
